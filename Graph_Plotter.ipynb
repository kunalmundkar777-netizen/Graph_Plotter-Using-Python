```python
# ============================================
# Project: Graph Plotter using Python
# Author : Kunal Jaydas Mundkar
# ============================================

# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 400)

# -------------------------------
# Quadratic Function
# -------------------------------
y1 = x**2

plt.figure(figsize=(6,4))
plt.plot(x, y1)
plt.title("Graph of y = x^2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()


# -------------------------------
# Linear Function
# -------------------------------
y2 = 2*x + 3

plt.figure(figsize=(6,4))
plt.plot(x, y2)
plt.title("Graph of y = 2x + 3")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()


# -------------------------------
# Sine Function
# -------------------------------
y3 = np.sin(x)

plt.figure(figsize=(6,4))
plt.plot(x, y3)
plt.title("Graph of y = sin(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()


# -------------------------------
# Cosine Function
# -------------------------------
y4 = np.cos(x)

plt.figure(figsize=(6,4))
plt.plot(x, y4)
plt.title("Graph of y = cos(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()

print("Graphs plotted successfully!")
```
